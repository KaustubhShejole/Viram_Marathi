```python
from datasets import load_dataset

ds = load_dataset("thenlpresearcher/test_data_marathi")
df = ds["test"].to_pandas()
```

```python
# rename columns
df = df.rename(columns={
    "gemini_out": "gt",
    "gt_marathi": "indictrans2_en_indic_pe",
    "cfilt_out": "cfilt_model"
})

# keep only columns that must survive
df = df[["sent_written", "sent_meant", "gt", "indictrans2_en_indic_pe",
         "cfilt_model", "punct_type"]]
```

```python
df.head()
```

```python
from datasets import Dataset, DatasetDict

# Convert pandas → HF Dataset
test_ds = Dataset.from_pandas(df, preserve_index=False)

# Create a DatasetDict with only a test split
ds_dict = DatasetDict({
    "test": test_ds
})

# Push to Hub with the proper structure
ds_dict.push_to_hub("thenlpresearcher/test_data_marathi")
```
