```python
from bleurt import score
import numpy as np

def compute_scores(preds, references):
    scores = scorer.score(candidates=preds, references=references)
    
    # Return corpus-level BLEURT as average
    return float(np.mean(scores))
```

```python
import os
print(os.getcwd())
```

```python
scorer = score.BleurtScorer("../../../Approach2/shalaka/bleurt/bleurt_checkpoints/base-128/bleurt-base-128")
```

```python
bleurt = {}
```

## Original

```python
file_name = "original_outputs.csv"
mode="original"
```

```python
import pandas as pd
df = pd.read_csv(file_name)
df.head()
```

```python
translations = df['prediction_mar']
ref_gem = df['gt_mar']
```

```python
bleurt_score = compute_scores(translations, ref_gem)
bleurt[mode] = bleurt_score
print(f"BleuRT Score for '{mode}' Model: ", bleurt_score)
```
