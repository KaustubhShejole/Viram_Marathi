```python
import pandas as pd
from evaluate import load

bleu = load("sacrebleu")
chrf = load("chrf")
cometm = load("comet")
```

```python
def compute_scores(preds, ref, src_sentences):
    """
    Compute BLEU and chrF++ scores using all three references for each sentence.
    """
    bleu_score = bleu.compute(predictions=preds, references=ref)["score"]
    chrf_score = chrf.compute(predictions=preds, references=ref)["score"]
    chrf2_score = chrf.compute(predictions=preds, references=ref, char_order=6, word_order=2,  beta=2)["score"]
    comet_score = cometm.compute(sources=src_sentences, predictions=preds, references=ref)["mean_score"]
    return bleu_score, chrf_score, chrf2_score, comet_score
```

## Complete Scores

```python
import os
import pandas as pd

# Your scoring function (assumed already defined)
# from your_module import compute_scores  

# Get all CSV files in the current directory
csv_files = [f for f in os.listdir('.') if f.endswith('.csv')]

# Sort CSV files based on the part after the first underscore
def sort_key(filename):
    # Split on '_' and take everything after the first '_'
    parts = filename.split('_', 1)
    return parts[1] if len(parts) > 1 else ''

csv_files.sort(key=sort_key)

# Process each CSV
for file_name in csv_files:
    print(f"Processing {file_name}...")
    
    df = pd.read_csv(file_name)
    
    src_sentences = df['Sentence']
    ref_gem = df['gt']
    translations = df['Response']
    
    bleu_score, chrf_score, chrf2_score, comet_score = compute_scores(translations, ref_gem, src_sentences)
    
    print(f"\n===== FINAL METRICS FOR '{file_name}'=====")
    print(f"BLEU: {bleu_score:.2f}, chrF++: {chrf_score:.2f}, chrF2++: {chrf2_score:.2f}, COMET: {comet_score:.4f}")
```
