```python
from bleurt import score
import numpy as np

def compute_scores(preds, references):
    scores = scorer.score(candidates=preds, references=references)
    
    # Return corpus-level BLEURT as average
    return float(np.mean(scores))
```

```python
import os
import pandas as pd
# CSV folder is current directory
csv_folder = "."

# Collect results
results = []

for file_name in os.listdir(csv_folder):
    if file_name.endswith(".csv"):
        file_path = os.path.join(csv_folder, file_name)
        df = pd.read_csv(file_path)

        predictions =  [str(x) for x in df['Response'].tolist()]
        references = [str(x) for x in df['gt'].tolist()]

        # Compute translation scores
        scores = compute_scores(translations, ref_gem)

        # Append results with file info
        results.append({
            "file": file_name,
            "strategy": file_name.split('_')[1],  # part after first '_'
            "scores": scores
        })

# Sort results by strategy (the word after the first '_')
results_sorted = sorted(results, key=lambda x: x['strategy'])

# Print sorted results
for r in results_sorted:
    print(f"File: {r['file']}, Strategy: {r['strategy']}, Scores: {r['scores']}")
```
