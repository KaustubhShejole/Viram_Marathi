```python
import pandas as pd
import os

# Path to your Excel file
excel_path = "responses_updated.xlsx"

# Folder names to match against
folders = [
    "few_cot",
    "few_zero",
    "original",
    "zero_cot",
    "zero_zero"
]

# Load the Excel file
xls = pd.ExcelFile(excel_path)
print(xls.sheet_names)
for sheet_name in xls.sheet_names:
    for folder in folders:
        if sheet_name.endswith(folder):
            # Ensure folder exists
            os.makedirs(folder, exist_ok=True)

            # Read sheet
            df = pd.read_excel(xls, sheet_name=sheet_name)

            # Output path
            output_path = os.path.join(folder, f"{sheet_name}.csv")

            # Save sheet as its own Excel file
            df.to_csv(output_path, index=False)

            print(f"Saved sheet '{sheet_name}' → {output_path}")
            break
```
